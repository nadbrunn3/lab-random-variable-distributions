# Lab | Random variable distributions

### Instructions

1. Get the numerical variables from our dataset.
2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
df_num = df.select_dtypes(include = np.number)
```

```python
for i in df_num:
    sns.distplot(df_num[i])
    plt.show()
    
# Customer_Lifetime_Value = Exponential 
# Income = hard to say
# Monthly_Premium_Auto = more exponential
# Month_Since_Last_Claim = hard to say
# Month_Policy_Inception = a little bit normal distributed
# Total_Claim_Amount = Exponential
```


### LOG TRANSFORMER:


```python
data_log = np.log(df['Total Claim Amount'])

fig, axs = plt.subplots(1,4)
axs[0].hist(df['Total Claim Amount'])
axs[0].set_title('Original Data')
axs[1].hist(data_log)
axs[1].set_title('Log Transformed Data')


```

```python
data_log = np.log(df['Monthly Premium Auto'])

fig, axs = plt.subplots(1,2)
axs[0].hist(df['Monthly Premium Auto'])
axs[0].set_title('Original Data')
axs[1].hist(data_log)
axs[1].set_title('Log Transformed Data')
```

```python
data_log = np.log(df['Customer Lifetime Value'])

fig, axs = plt.subplots(1,2)
axs[0].hist(df['Customer Lifetime Value'])
axs[0].set_title('Original Data')
axs[1].hist(data_log)
axs[1].set_title('Log Transformed Data')
```
